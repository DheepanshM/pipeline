```python
import pandas as pd
```

```python
df_transaction=pd.read_csv("transaction_data.csv")
df_us_customer=pd.read_csv("us_customer_data 1.csv")
```

```python
df_transaction=df_transaction.drop_duplicates()
df_us_customer=df_us_customer.drop_duplicates()
```

```python
df_us_customer['email']=df_us_customer['email'].fillna(df_us_customer['name'].str.replace(' ','').str.lower()+'@example.com')
```

```python
df_transaction['amount']=df_transaction['amount'].fillna(df_transaction['amount'].mean)
```

```python
df_us_customer['country_code']=df_us_customer['address'].str[-8:-6]
```

```python
df_us_customer['country_code'].unique()
```

```python
df_us_customer['phone']=df_us_customer['phone'].fillna(0)
df_us_customer['phone']=df_us_customer['phone'].astype(str)
```

```python
def standard_phone(df_us_customer):
    si=len(df_us_customer['phone'])
    for i in range(si):
        if len(df_us_customer['phone'][i])>12 or len(df_us_customer['phone'][i])<12:
            df_us_customer['phone'][i]='0'
        elif df_us_customer['country_code'][i]=='IN':
            df_us_customer['phone'][i]='+91 '+str (df_us_customer['phone'][i])
        elif df_us_customer['country_code'][i]=='PW':
            df_us_customer['phone'][i]='+680 '+str (df_us_customer['phone'][i])
        elif df_us_customer['country_code'][i]=='FM':
            df_us_customer['phone'][i]='+691 '+str (df_us_customer['phone'][i])
        elif df_us_customer['country_code'][i]=='AE':
            df_us_customer['phone'][i]='+971 '+str (df_us_customer['phone'][i])
        else:
            df_us_customer['phone'][i]='+1 '+str (df_us_customer['phone'][i])
        
        
    return df_us_customer

standard_phone(df_us_customer)
```
